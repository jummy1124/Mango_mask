<h3>Inport required package</h3>

```python
import matplotlib.pyplot as plt
import numpy as np
from keras.models import load_model
from keras.preprocessing import image
```

<h3>Set tensorflow</h3>

```python
import tensorflow as tf

config = tf.ConfigProto()
config.gpu_options.allow_growth = True
session = tf.InteractiveSession(config=config)
```

<h3>Load test image and view</h3>

```python
test_image=image.load_img('./test_image.jpg',target_size=(512,512,3))
img=image.img_to_array(test_image)
img=img/255.0
plt.imshow(img)
```

<h3>Predict mango image and view</h3>

```python
model=load_model('unet.h5')
new_img=np.expand_dims(img, axis=0)
predict=model.predict(new_img)
print(predict.shape)

mango=np.zeros(shape=(512,512,1))
for x in range(512):
    for y in range(512):
        mango[x,y,0]=np.argmax(predict[0][x][y],axis=0)
plt.imshow(mango[:,:,0])
```

<h3>Remove background of original mango image</h3>

```python
new_mango=np.ones(shape=(512,512,3))
for x in range(512):
    for y in range(512):
        if mango[x,y,0]==1:   
            new_mango[x,y,:]=img[x,y,:]
plt.imshow(new_mango)
```
